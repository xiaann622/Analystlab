## Task 2: Data Cleaning

### 2.1 Missing Values

```python
missing = df.isnull().sum()
missing_pct = (missing / len(df) * 100).round(2)
missing_summary = pd.DataFrame({'missing_count': missing, 'missing_pct': missing_pct})
missing_summary = missing_summary[missing_summary['missing_count'] > 0].sort_values('missing_count', ascending=False)
missing_summary
```

**Findings & decisions:**

- **`director`** (typically the largest gap, ~30%): Often genuinely absent from Netflix's own metadata (common for TV shows with rotating directors per episode). Decision: **fill with `"Not Specified"`** rather than drop — dropping would remove ~30% of the catalog, destroying data needed for every other analysis.
- **`cast`** (~9%): Same reasoning — **fill with `"Not Specified"`**.
- **`country`** (~9%): Genuinely unknown production country. Decision: **fill with `"Not Specified"`**; it's excluded automatically from country-level analyses since it doesn't match a real country name.
- **`date_added`** (a small number of rows): This is a date Netflix's own catalog system stamped; there's no reliable way to reconstruct it. Decision: **drop only those rows** for analyses that require `date_added` (e.g., "content added by year"), but keep them for everything else — dropping them wholesale would lose valid title/genre/rating data.
- **`rating`** (a handful of rows): Decision: **fill with `"Not Rated"`**, a legitimate real-world category already used by other rating systems.
- **`duration`** (a handful of rows): Decision: **drop these specific rows** only when duration is required for analysis, since duration cannot be meaningfully guessed.


```python
df['director'] = df['director'].fillna('Not Specified')
df['cast'] = df['cast'].fillna('Not Specified')
df['country'] = df['country'].fillna('Not Specified')
df['rating'] = df['rating'].fillna('Not Rated')
# date_added and duration: leave as NaN for now, handled explicitly downstream

print("Remaining missing values:")
print(df.isnull().sum()[df.isnull().sum() > 0])
```

### 2.2 Duplicate Records

```python
n_duplicates_all_cols = df.duplicated().sum()
# Also check duplicates on the natural identifier + title, in case show_id
# itself was regenerated but the underlying title/type is a re-listing
n_duplicates_title = df.duplicated(subset=['title', 'type', 'country', 'release_year']).sum()

print("Exact duplicate rows (all columns):", n_duplicates_all_cols)
print("Duplicate title+type+country+release_year combos:", n_duplicates_title)
```

```python
before = len(df)
df = df.drop_duplicates()
df = df.drop_duplicates(subset=['title', 'type', 'country', 'release_year'], keep='first')
after = len(df)
print(f"Duplicates removed: {before - after}")
print(f"Rows before: {before} | Rows after: {after}")
```

### 2.3 Standardization

```python
# --- Column names ---
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
print(df.columns.tolist())
```

```python
# --- Date formats ---
df['date_added'] = df['date_added'].astype(str).str.strip()
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
print("Unparseable / missing date_added after conversion:", df['date_added'].isnull().sum())
```

```python
# --- Text formatting consistency ---
df['type'] = df['type'].str.strip().str.title()          # Movie / Tv Show -> Movie / TV Show below
df['type'] = df['type'].replace({'Tv Show': 'TV Show'})
df['title'] = df['title'].str.strip()
df['country'] = df['country'].str.strip()
df['rating'] = df['rating'].str.strip().str.upper()
df['listed_in'] = df['listed_in'].str.strip()
df[['type','title','country','rating']].head()
```

```python
# --- Data types: release_year should be int, date_added extracted parts for EDA ---
df['release_year'] = df['release_year'].astype(int)
df['year_added'] = df['date_added'].dt.year
df['month_added'] = df['date_added'].dt.month_name()
df.dtypes
```

### 2.4 Data Validation — Invalid Values, Inconsistencies, Outliers

```python
# Split duration into a numeric value + unit, since it mixes "X min" (movies)
# and "X Season(s)" (TV shows) in one text field — not directly usable as a number.
df['duration_value'] = df['duration'].str.extract(r'(\d+)').astype(float)
df['duration_unit'] = df['duration'].str.extract(r'([A-Za-z]+)')

print("Rows with unparseable duration:", df['duration_value'].isnull().sum())
df[['type','duration','duration_value','duration_unit']].head()
```

```python
# Invalid values: release_year should be realistic (Netflix data starts well
# after film existed, e.g. 1925+, and can't be in the future beyond data collection)
current_year_ceiling = 2026
invalid_years = df[(df['release_year'] < 1900) | (df['release_year'] > current_year_ceiling)]
print("Rows with implausible release_year:", len(invalid_years))
```

```python
# Inconsistent country entries: 'country' is comma-separated multi-value —
# for country-level analysis, we standardize by taking the FIRST listed country
# as the primary production country (documented assumption).
df['primary_country'] = df['country'].apply(lambda x: x.split(',')[0].strip() if x != 'Not Specified' else x)
df['primary_country'].value_counts().head(10)
```

```python
# Inconsistent genre entries: 'listed_in' is also comma-separated multi-label.
# We keep the raw column for reference and additionally create an exploded
# version for genre-level EDA (one row per title-genre pair).
genre_exploded = df.assign(genre=df['listed_in'].str.split(',')).explode('genre')
genre_exploded['genre'] = genre_exploded['genre'].str.strip()
print("Exploded genre rows:", len(genre_exploded), "vs original rows:", len(df))
```

```python
# Outlier check on duration (numeric part), separated by type since
# 'minutes' and 'seasons' are not comparable scales
for t in ['Movie', 'TV Show']:
    sub = df[(df['type'] == t) & df['duration_value'].notnull()]['duration_value']
    q1, q3 = sub.quantile([0.25, 0.75])
    iqr = q3 - q1
    lower, upper = q1 - 1.5*iqr, q3 + 1.5*iqr
    n_outliers = ((sub < lower) | (sub > upper)).sum()
    print(f"{t}: IQR bounds [{lower:.1f}, {upper:.1f}] -> {n_outliers} potential outliers "
          f"({n_outliers/len(sub)*100:.1f}% of {t.lower()}s)")
```

### 2.5 Cleaning Summary

```python
cleaning_summary = pd.DataFrame([
    ["Missing director (~30%)", "Filled with 'Not Specified'"],
    ["Missing cast (~9%)", "Filled with 'Not Specified'"],
    ["Missing country (~9%)", "Filled with 'Not Specified'"],
    ["Missing rating (small #)", "Filled with 'Not Rated'"],
    ["Missing date_added / duration (small #)", "Left as NaN; excluded row-by-row only from analyses that need that field"],
    ["Exact duplicate rows", "Removed"],
    ["Duplicate title+type+country+release_year", "Removed (kept first occurrence)"],
    ["Column names", "Standardized to lower_snake_case"],
    ["date_added as text", "Converted to datetime; year_added/month_added derived"],
    ["'Tv Show' inconsistent casing", "Standardized to 'TV Show'"],
    ["duration mixed units ('90 min' / '2 Seasons')", "Split into duration_value (numeric) + duration_unit"],
    ["Multi-value country field", "Standardized via primary_country (first listed country)"],
    ["Multi-value listed_in (genre) field", "Exploded into a separate long-format table for genre-level analysis"],
    ["Implausible release_year", "Checked via bounds (1900–2026); none removed unless found"],
], columns=["Issue Found", "Action Taken"])
cleaning_summary
```

## Task 4: Data Visualization

```python
# Chart 1: Movies vs TV Shows (Pie Chart)
plt.figure(figsize=(6,6))
plt.pie(type_counts, labels=type_counts.index, autopct='%1.1f%%',
        colors=['crimson', 'navy'], startangle=90)
plt.title('Movies vs TV Shows on Netflix')
plt.tight_layout()
plt.show()
```

**Finding:** Movies make up a clear majority of the catalog relative to TV Shows, reflecting Netflix's historically larger film library.

```python
# Chart 2: Content added by year (Line Chart)
plt.figure(figsize=(10,5))
content_by_year.plot(kind='line', marker='o', color='darkred')
plt.title('Number of Titles Added to Netflix by Year')
plt.xlabel('Year Added')
plt.ylabel('Number of Titles')
plt.tight_layout()
plt.show()
```

**Finding:** Content additions ramp up sharply in the mid-2010s and peak around 2018–2019, consistent with Netflix's global content-investment expansion, before tapering in the final (partial) year of data collection.

```python
# Chart 3: Top 10 content-producing countries (Bar Chart)
plt.figure(figsize=(10,6))
top_countries.sort_values().plot(kind='barh', color='teal')
plt.title('Top 10 Content-Producing Countries')
plt.xlabel('Number of Titles')
plt.ylabel('Country')
plt.tight_layout()
plt.show()
```

**Finding:** The United States dominates as the primary production country by a wide margin, with India and the UK typically rounding out the top few — reflecting Netflix's original content hubs.

```python
# Chart 4: Most common ratings (Bar Chart)
plt.figure(figsize=(10,6))
top_ratings.plot(kind='bar', color='darkorange')
plt.title('Most Common Content Ratings')
plt.xlabel('Rating')
plt.ylabel('Number of Titles')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

**Finding:** Mature-audience ratings (e.g. TV-MA, TV-14) dominate the catalog, suggesting Netflix's content mix skews toward adult/teen audiences over younger children's content.

```python
# Chart 5: Movie duration distribution (Histogram)
movie_durations = df[df['type'] == 'Movie']['duration_value'].dropna()
plt.figure(figsize=(10,5))
plt.hist(movie_durations, bins=40, color='mediumseagreen', edgecolor='black')
plt.title('Distribution of Movie Durations (Minutes)')
plt.xlabel('Duration (minutes)')
plt.ylabel('Number of Movies')
plt.tight_layout()
plt.show()
```

**Finding:** Movie runtimes cluster tightly around 90–110 minutes, the conventional feature-film length, with a smaller number of much shorter or longer outliers.

```python
# Chart 6 (bonus): TV Show seasons distribution (Box Plot)
tv_seasons = df[df['type'] == 'TV Show']['duration_value'].dropna()
plt.figure(figsize=(8,5))
sns.boxplot(x=tv_seasons, color='plum')
plt.title('Distribution of TV Show Season Counts')
plt.xlabel('Number of Seasons')
plt.tight_layout()
plt.show()
```

**Finding:** Most TV shows on Netflix have just 1–2 seasons, with a long tail of a small number of long-running series flagged as outliers on the box plot.

## Task 5: Insights

1. **Movies dominate the catalog over TV Shows** (see *Chart 1*), roughly a 2:1 ratio — Netflix's library is still primarily film-driven despite its reputation for original series.
2. **Content acquisition accelerated sharply from the mid-2010s, peaking around 2018–2019** (see *Chart 2*), aligning with Netflix's well-documented global expansion and original-content investment period.
3. **The United States is by far the largest content-producing country** (see *Chart 3*), with a small number of other countries (typically India, UK) forming a distant second tier — indicating both a US-centric library and a limited-but-growing set of international content hubs.
4. **The catalog skews toward mature audiences**: TV-MA and TV-14 are the most common ratings (see *Chart 4*), suggesting Netflix's content strategy prioritizes teen/adult viewers over young children.
5. **Most TV shows are short-lived** (1–2 seasons; see the season box plot), while movies cluster tightly around the standard ~90–110 minute runtime (see *Chart 5*) — together suggesting Netflix favors a high volume of shorter series/limited runs over sustained long-running shows.

```python

missing_values = data.isnull().sum()


missing_values = missing_values[missing_values > 0]

print("Columns with missing values and their counts:")
display(missing_values)
```

```python
# Calculate the percentage of missing values
total_rows = len(data)
missing_percentage = (missing_values / total_rows) * 100

print("Percentage of missing values per column:")
display(missing_percentage.sort_values(ascending=False))
```

### Handling Missing Values

**Strategy:**

1.  **CustomerID:** As `CustomerID` is a crucial identifier for customer-centric analysis and approximately 25% of its values are missing, removing these rows is the most appropriate approach to avoid skewed analysis.
2.  **Description:** With a very small percentage (0.27%) of missing values, filling `Description` with 'Unknown' ensures all transaction records are preserved, which can be useful for broader analysis not directly dependent on product descriptions.

```python
# 1. Handling missing CustomerID values
before_drop_rows = len(data)
data.dropna(subset=['CustomerID'], inplace=True)
after_drop_rows = len(data)
dropped_customer_id_rows = before_drop_rows - after_drop_rows
print(f"Removed {dropped_customer_id_rows} rows due to missing CustomerID.")

# 2. Handling missing Description values
data['Description'].fillna('Unknown', inplace=True)
print("Filled missing 'Description' values with 'Unknown'.")
```

### Verification of Missing Values

Let's verify that there are no longer any missing values in the `CustomerID` and `Description` columns, and check for any other remaining missing values.

```python
missing_values_after_handling = data.isnull().sum()
missing_values_after_handling = missing_values_after_handling[missing_values_after_handling > 0]

if missing_values_after_handling.empty:
    print("No more missing values in the dataset.")
else:
    print("Remaining columns with missing values:")
    display(missing_values_after_handling)
```

### Duplicate Records

Let's identify and remove any duplicate rows in the dataset. A duplicate row means an identical entry across all columns. Removing them ensures that each transaction line is unique and prevents overcounting in analyses.

```python
# Identify duplicate rows
duplicate_rows = data.duplicated().sum()
print(f"Number of duplicate rows found: {duplicate_rows}")
```

```python
# Remove duplicate rows
before_drop_duplicates = len(data)
data.drop_duplicates(inplace=True)
after_drop_duplicates = len(data)

duplicates_removed = before_drop_duplicates - after_drop_duplicates
print(f"Number of duplicate rows removed: {duplicates_removed}")
print(f"Number of rows after removing duplicates: {len(data)}")
```
